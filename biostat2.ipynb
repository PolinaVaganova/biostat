```python
import pandas as pd
import os
import numpy as np

path_to_dir = r"C:\Users\lenovo\Documents\6 семестр\статистика\спбгу"

# task 6.1: find xi-M and (xi-M)^2
df = pd.read_excel(os.path.join(path_to_dir, "input1.1.xlsx"))
df["xi-M"] = df["xi"] - df["xi"].mean()
df["(xi-M)^2"] = (df["xi"] - df["xi"].mean()) ** 2
df.round(2)
```

```python
# task 6.2: make function to find statistics 
def find_stat(df):
    statistics = {"n": len(df["xi"]),
    "M": df["xi"].mean(),
    "Me": df["xi"].median(),
    "Mo": df["xi"].mode()[0],
    "min": df["xi"].min(),
    "max": df["xi"].max(),
    "R": df["xi"].max()-df["xi"].min(),
    "s2": df["xi"].var(),
    "SD": df["xi"].std(),
    "As": df["xi"].skew(),
    "Ex": df["xi"].kurtosis()}
    
    stat_df = pd.DataFrame.from_dict(statistics, orient='index').astype(float)
    return stat_df.round(1)

find_stat(df)
```

```python
# task 6.3: weighted series

df = pd.read_excel(os.path.join(path_to_dir, "input1.2.xlsx"))
df["fi*xi"] = df["fi"] * df["xi"] 
df["(xi - M)"] = df["xi"]-df["xi"].mean()
df["fi * (xi - M)^2"] = df["fi"] * df["(xi - M)"]**2
df["fi * (xi - M)^3"] = df["fi"] * df["(xi - M)"]**3
df["fi * (xi - M)^4"] = df["fi"] * df["(xi - M)"]**4
df
```

```python
print("сумма xi", df["xi"].sum())
print("сумма (xi - M)^2", df["fi * (xi - M)^2"].sum())
print("сумма (xi - M)^3", df["fi * (xi - M)^3"].sum())
print("сумма (xi - M)^4", df["fi * (xi - M)^4"].sum())
```

```python
# task 6.4: statistics for weighted series

df_1=df["интервалы"].tolist()
df_2=[]
for i in range(len(df_1)):
    elemnt = (int(df_1[i].split(" ")[0]) + int(df_1[i].split(" ")[2]))//2
    df_2.append(elemnt)
df["xi"] = df_2
df = df.loc[df.index.repeat(df["fi"])]
df.reset_index(drop= True , inplace= True )

find_stat(df)
```

```python
# task 10: Для группы рыб рассчитайте доли 

df = pd.read_excel(os.path.join(path_to_dir, "input1.10.xlsx"))
sampl_size = len(df)

# а) самцов:
print(round((len(df[(df["пол"] == "самец")]) / sampl_size), 2))

# б) рыб с красной окраской,
print(round((len(df[(df["окраска"] == "красный")]) / sampl_size), 2))


# в) половозрелых рыб среди самок и самцов,
print(round((len(df[(df["половозрелость"] == 1)]) / sampl_size), 2))

# г) рыб крупнее 45 см,
print(round((len(df[(df["длина, см"] >= 45)]) / sampl_size), 2))

# д) самок от 40 до 50 см.
print(round((len(df[(df["пол"] == "самка") & (df["длина, см"] <= 50) & (df["длина, см"] >= 40)]) / sampl_size), 2))
```
